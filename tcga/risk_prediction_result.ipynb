```python
%reset -f
%matplotlib inline
import pandas as pd
from scipy.stats import ranksums
```

```python
result = pd.read_excel("./result.xlsx")
mean_res = result.groupby(['Cancer', 'Model']).mean()
mean_res.columns = ['C-index mean', 'ibs mean']
std_res = result.groupby(['Cancer', 'Model']).std()
std_res.columns = ['C-index std', 'ibs std']

final_result = pd.concat((mean_res, std_res), axis=1)
final_result.to_excel("./risk_prediction_results.xlsx")
final_result
```

The p-value of less than 0.05 indicates that the test rejects the null hypothesis that the two sets of measurements are drawn from the same distribution at the 5% significance level.

```python
cancers = ["GBM", "KIRC", "BRCA"]
for cancer in cancers:
    c_idx_bina = result[(result.Cancer == cancer) & (result.Model == 'Binacox')]["C-index"]
    c_idx_rsf = result[(result.Cancer == cancer) & (result.Model == 'RSF')]["C-index"]
    c_idx_coxb = result[(result.Cancer == cancer) & (result.Model == 'CoxBoost')]["C-index"]
    
    ibs_bina = result[(result.Cancer == cancer) & (result.Model == 'Binacox')]["ibs"]
    ibs_rsf = result[(result.Cancer == cancer) & (result.Model == 'RSF')]["ibs"]
    ibs_coxb = result[(result.Cancer == cancer) & (result.Model == 'CoxBoost')]["ibs"]
    
    p_val = ranksums(c_idx_bina, c_idx_rsf).pvalue
    print("\nRank test %s C-index Binacox VS. RSF: %.2E" % (cancer, p_val))
    p_val = ranksums(c_idx_bina, c_idx_coxb).pvalue
    print("Rank test %s C-index Binacox VS. CoxBoost: %.2E" % (cancer, p_val))
    p_val = ranksums(ibs_bina, ibs_rsf).pvalue
    print("Rank test %s IBS Binacox VS. RSF: %.2E" % (cancer, p_val))
    p_val = ranksums(ibs_bina, ibs_coxb).pvalue
    print("Rank test %s IBS Binacox VS. CoxBoost: %.2E" % (cancer, p_val))
```
